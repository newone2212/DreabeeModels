```python
import spacy
from spacy.language import Language
from langdetect import detect, DetectorFactory
from langdetect.lang_detect_exception import LangDetectException

# To make the language detection deterministic
DetectorFactory.seed = 0

# Define custom language detection component
@Language.component("language_detector")
def language_detector(doc):
    try:
        lang = detect(doc.text)
    except LangDetectException:
        lang = "unknown"
    doc._.language = lang
    return doc

# Load spaCy model
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("language_detector", last=True)

# Register the custom extension on the Doc object
from spacy.tokens import Doc
Doc.set_extension("language", default="unknown", force=True)

# Function to detect language
def detect_language(text):
    doc = nlp(text)
    return doc._.language

# Test the language detection
text = "In the midst of a bustling city, where the sound of honking cars and the chatter of pedestrians created a constant hum, a small bird, perched on the edge of a rusted metal balcony, observed the world below with curious, beady eyes, taking in the vibrant colors of the street market, where vendors shouted their wares, the smell of freshly baked bread wafted through the air, mingling with the scent of blooming flowers from a nearby florist, while children laughed and played, their joyful voices rising above the din, and the occasional breeze carried with it the distant echo of a musician strumming a guitar, creating a melody that seemed to dance through the air, weaving in and out of the chaotic symphony of the city, until finally, the bird, with a sudden burst of energy, spread its wings and took flight, soaring above the rooftops, where the noise began to fade, replaced by the gentle rustling of leaves in the wind, and the vast, open sky stretched out before it, offering a sense of freedom and tranquility that could only be found far above the frenzied pace of life below."
print(f"Text: {text}\nDetected language: {detect_language(text)}")

text = "Ceci est une phrase de test."
print(f"Text: {text}\nDetected language: {detect_language(text)}")

text = "Dies ist ein Testsatz."
print(f"Text: {text}\nDetected language: {detect_language(text)}")

text = "这是一个测试句子。"
print(f"Text: {text}\nDetected language: {detect_language(text)}")

# # Example using spaCy's NLP pipeline
# doc = nlp("This is a sentence.")
# print([(w.text, w.pos_) for w in doc])
```

```python
# load_language_detector.py

import spacy
import langid
import os

# Function to detect language using langid
def detect_language(text):
    lang_code, _ = langid.classify(text)
    
    lang_code_to_name = {
        'en': 'English',
        'fr': 'French',
        'de': 'German',
        'zh': 'Chinese',
        'so': 'Somali',
        'hi': 'Hindi',  # Added Hindi for your specific example
        'es': 'Spanish',
        'ar': 'Arabic',
        # Add more language mappings as needed
    }
    
    return lang_code_to_name.get(lang_code, "Unknown")

# Load the saved spaCy model
load_dir = r"ml_models\models\language_model"
nlp = spacy.load(load_dir)


# Example texts
texts = [
    "This is a test sentence.",
    "Ceci est une phrase de test.",
    "Dies ist ein Testsatz.",
    "这是一个测试句子。",
    "ya kya kar raha ho.",
    "mere naam parth hai"
]

# Detect language for each text
for text in texts:
    print(f"Text: {text}\nDetected language: {detect_language(text)}\n")

# Example using spaCy's NLP pipeline
doc = nlp("This is a sentence.")
print([(w.text, w.pos_) for w in doc])
```

```python
from optimum.onnxruntime import ORTModelForSequenceClassification
from transformers import AutoTokenizer, pipeline

# Specify the ONNX file name
model_file_name = "model.onnx"  # Change this if you want to use a different ONNX file

# Load the tokenizer and the model
tokenizer = AutoTokenizer.from_pretrained("laiyer/xlm-roberta-base-language-detection-onnx")
model = ORTModelForSequenceClassification.from_pretrained(
    "laiyer/xlm-roberta-base-language-detection-onnx",
    file_name=model_file_name
)

# Create a pipeline for text classification
classifier = pipeline(
    task="text-classification",
    model=model,
    tokenizer=tokenizer,
    top_k=None,
)

# Define the text to classify
text = "Mai parth hu"

# Get the classification result
classifier_output = classifier(text)
print(classifier_output)
```
